```python
reverse_nuts_inv = {j:i for i,j in nuts_inv.items()} #reverse keys and values; 
reverse_nuts_inv
```

```python
# jde dict to cad dict:
#e.g {185671:185680,  24107929:185536}
def jde2cad(sample):
    return {reverse_nuts_inv.get(j): reverse_nuts_inv.get(k) for j,k in sample.items()}
```

```python
jde2cad({185671:185680})
```
